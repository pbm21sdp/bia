```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_gaussian_quantiles
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X1, y1 = make_gaussian_quantiles(cov=7., n_samples=250, n_features=2, n_classes=3, random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.45, random_state=42)

svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm_clf', SVC(kernel='poly', degree=3, coef0=10, C=5))])

svm_pipeline.fit(X_train, y_train)
y_pred = svm_pipeline.predict(X_test)
accuracy_score(y_test, y_pred)

X_2D = np.c_[X1, X1**2]
y = y1

plt.figure(figsize=(10, 3))

plt.subplot(121)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.plot(X1[:, 0][y==0], np.zeros(X1[:, 0][y==0].shape[0]), "bs")
plt.plot(X1[:, 0][y==1], np.zeros(X1[:, 0][y==1].shape[0]), "g^")
plt.gca().get_yaxis().set_ticks([])
plt.xlabel(r"$x_1$", fontsize=20)
plt.axis([-4.5, 4.5, -0.2, 0.2])

plt.subplot(122)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(X_2D[:, 0][y==0], X_2D[:, 1][y==0], "bs")
plt.plot(X_2D[:, 0][y==1], X_2D[:, 1][y==1], "g^")
plt.xlabel(r"$x_1$", fontsize=20)
plt.ylabel(r"nu stiu$", fontsize=20, rotation=0)
plt.gca().get_yaxis().set_ticks([0, 4, 8, 12, 16])
plt.plot([-10, 10], [6.5, 6.5], "r--", linewidth=3)
plt.axis([-10, 10, -10, 17])

plt.show()
```
